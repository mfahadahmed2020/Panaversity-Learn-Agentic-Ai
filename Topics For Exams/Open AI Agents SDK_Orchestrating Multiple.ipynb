## Open AI Agents SDK_Orchestrating Multiple Agents /  اوپن اےآئی ایجنٹس ایس ڈی کے آرکیسٹریٹنگ متعددایجنٹس 

Orchestration refers to the flow of agents in your app. Which agents run, in what order, and how do they decide what happens next? There are two main ways to orchestrate agents:

1. Allowing the LLM to make decisions: this uses the intelligence of an LLM to plan, reason 
   and decide on what steps to take based on that.

2. Orchestrating via code: determining the flow of agents via your code.

You can mix and match these patterns. Each has their own tradeoffs, described below.

آرکیسٹریشن سے مراد آپ کی ایپ میں ایجنٹوں کا بہاؤ ہے۔ کون سے ایجنٹ چلتے ہیں، کس ترتیب میں، اور وہ کیسے فیصلہ کرتے ہیں کہ آگے کیا ہوتا ہے؟ ایجنٹوں کو آرکیسٹریٹ کرنے کے دو اہم طریقے ہیں:

1. ایل ایل ایم کو فیصلے کرنے کی اجازت دینا: یہ منصوبہ بندی کے لیے ایل ایل ایم کی ذہانت کا استعمال کرتا ہے، وجہ 
اور فیصلہ کریں کہ اس کی بنیاد پر کیا اقدامات کیے جائیں۔

2. کوڈ کے ذریعے آرکیسٹریٹنگ: اپنے کوڈ کے ذریعے ایجنٹوں کے بہاؤ کا تعین کرنا۔

آپ ان نمونوں کو مکس اور میچ کر سکتے ہیں۔ ہر ایک کی اپنی تجارت ہے، ذیل میں بیان کیا گیا ہے۔

Orchestrating via LLM

An agent is an LLM equipped with instructions, tools and handoffs. This means that given an open-ended task, the LLM can autonomously plan how it will tackle the task, using tools to take actions and acquire data, and using handoffs to delegate tasks to sub-agents. For example, a research agent could be equipped with tools like:

* Web search to find information online

* File search and retrieval to search through proprietary data and connections

* Computer use to take actions on a computer

* Code execution to do data analysis

* Handoffs to specialized agents that are great at planning, report writing and more.

ایل ایل ایم کے ذریعے آرکیسٹریٹنگ

ایجنٹ ایک LLM ہوتا ہے جو ہدایات، ٹولز اور ہینڈ آف سے لیس ہوتا ہے۔ اس کا مطلب یہ ہے کہ ایک کھلا کام دیا گیا ہے، LLM خود مختاری سے منصوبہ بنا سکتا ہے کہ وہ کس طرح کام سے نمٹائے گا، کارروائی کرنے اور ڈیٹا حاصل کرنے کے لیے ٹولز کا استعمال کرے گا، اور ذیلی ایجنٹوں کو کام سونپنے کے لیے ہینڈ آف کا استعمال کرے گا۔ مثال کے طور پر، ایک تحقیقی ایجنٹ ایسے آلات سے لیس ہو سکتا ہے جیسے:

* آن لائن معلومات تلاش کرنے کے لیے ویب تلاش کریں۔

* ملکیتی ڈیٹا اور کنکشن کے ذریعے تلاش کرنے کے لیے فائل کی تلاش اور بازیافت

* کمپیوٹر پر کارروائی کرنے کے لیے کمپیوٹر کا استعمال

* ڈیٹا کا تجزیہ کرنے کے لیے کوڈ پر عمل درآمد

* خصوصی ایجنٹوں کو ہینڈ آف جو منصوبہ بندی کرنے، رپورٹ لکھنے اور مزید بہت کچھ کرنے میں بہترین ہیں۔

This pattern is great when the task is open-ended and you want to rely on the intelligence of an LLM. The most important tactics here are:

1. Invest in good prompts. Make it clear what tools are available, how to use them, and what 
   parameters it must operate within.
2. Monitor your app and iterate on it. See where things go wrong, and iterate on your prompts.

3. Allow the agent to introspect and improve. For example, run it in a loop, and let it 
   critique itself; or, provide error messages and let it improve.

4. Have specialized agents that excel in one task, rather than having a general purpose agent 
   that is expected to be good at anything.

5. Invest in evals. This lets you train your agents to improve and get better at tasks.

یہ پیٹرن بہت اچھا ہے جب کام کھلا ہوا ہے اور آپ LLM کی ذہانت پر بھروسہ کرنا چاہتے ہیں۔ یہاں سب سے اہم حکمت عملی یہ ہیں:

1. اچھے اشارے میں سرمایہ کاری کریں۔ یہ واضح کریں کہ کون سے اوزار دستیاب ہیں، انہیں کیسے استعمال کیا جائے اور کیا کیا جائے۔ 
پیرامیٹرز کے اندر اندر کام کرنا ضروری ہے.
2. اپنی ایپ کی نگرانی کریں اور اس پر اعادہ کریں۔ دیکھیں کہ چیزیں کہاں غلط ہوتی ہیں، اور اپنے اشارے پر اعادہ کریں۔

3. ایجنٹ کو خود کا معائنہ کرنے اور بہتر بنانے کی اجازت دیں۔ مثال کے طور پر، اسے ایک لوپ میں چلائیں، اور اسے جانے دیں۔ 
خود تنقید؛ یا، غلطی کے پیغامات فراہم کریں اور اسے بہتر ہونے دیں۔

4. عام مقصد کے ایجنٹ کے بجائے خصوصی ایجنٹ رکھیں جو ایک کام میں مہارت حاصل کریں۔ 
اس سے کسی بھی چیز میں اچھا ہونے کی امید ہے۔

5. ایولز میں سرمایہ کاری کریں۔ یہ آپ کو اپنے ایجنٹوں کو کاموں میں بہتری اور بہتر بنانے کی تربیت دیتا ہے۔

Orchestrating via code

While orchestrating via LLM is powerful, orchestrating via code makes tasks more deterministic and predictable, in terms of speed, cost and performance. Common patterns here are:

* Using structured outputs to generate well formed data that you can inspect with your code.  
  For example, you might ask an agent to classify the task into a few categories, and then pick the next agent based on the category.

* Chaining multiple agents by transforming the output of one into the input of the next. You 
  can decompose a task like writing a blog post into a series of steps - do research, write an outline, write the blog post, critique it, and then improve it.

* Running the agent that performs the task in a while loop with an agent that evaluates and  
  provides feedback, until the evaluator says the output passes certain criteria.

* Running multiple agents in parallel, e.g. via Python primitives like asyncio.gather. This   
  is useful for speed when you have multiple tasks that don't depend on each other.

We have a number of examples in examples/agent_patterns.

کوڈ کے ذریعے آرکیسٹریٹنگ

جب کہ LLM کے ذریعے آرکیسٹریٹنگ طاقتور ہے، کوڈ کے ذریعے آرکیسٹریٹنگ رفتار، لاگت اور کارکردگی کے لحاظ سے کاموں کو زیادہ متعین اور پیش قیاسی بناتی ہے۔ یہاں عام پیٹرن ہیں:

* اچھی طرح سے تشکیل شدہ ڈیٹا تیار کرنے کے لیے سٹرکچرڈ آؤٹ پٹ کا استعمال کرنا جس کا آپ اپنے کوڈ سے معائنہ کر سکتے ہیں۔ 
مثال کے طور پر، آپ کسی ایجنٹ سے کام کو چند زمروں میں درجہ بندی کرنے کے لیے کہہ سکتے ہیں، اور پھر زمرہ کی بنیاد پر اگلا ایجنٹ چن سکتے ہیں۔

* ایک کے آؤٹ پٹ کو اگلے کے ان پٹ میں تبدیل کرکے متعدد ایجنٹوں کی زنجیر بنانا۔ آپ 
ایک بلاگ پوسٹ لکھنے جیسے کام کو کئی مراحل میں تحلیل کر سکتا ہے - تحقیق کریں، خاکہ لکھیں، بلاگ پوسٹ لکھیں، اس پر تنقید کریں، اور پھر اسے بہتر کریں۔

* اس ایجنٹ کو چلانا جو کام کو کچھ دیر میں انجام دیتا ہے ایک ایسے ایجنٹ کے ساتھ جو تشخیص کرتا ہے اور 
فیڈ بیک فراہم کرتا ہے، جب تک کہ جائزہ لینے والا یہ نہ کہے کہ آؤٹ پٹ کچھ معیارات سے گزر جاتا ہے۔

* متعدد ایجنٹوں کو متوازی طور پر چلانا، جیسے ازگر کے قدیم جیسے asyncio.gather کے ذریعے۔ یہ 
رفتار کے لیے مفید ہے جب آپ کے پاس متعدد کام ہیں جو ایک دوسرے پر انحصار نہیں کرتے ہیں۔
